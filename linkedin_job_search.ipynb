```python
# prompt: write a python code using selenium to search jobs on google.com website,matching the given skill set.the skill set is a list of strings that denote a skill

!pip install selenium
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin

import sys
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def search_jobs_with_skills(skills):
    """
    Searches for jobs on Google matching the given skillset.
    """

    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)

    wd.get("https://www.google.com/")

    search_box = wd.find_element(By.NAME, "q")
    search_query = "jobs " + " ".join(['"' + skill + '"' for skill in skills])
    search_box.send_keys(search_query)
    search_box.send_keys(Keys.RETURN)

    try:
        # Wait for search results to load
        WebDriverWait(wd, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".g"))
        )

        # Extract job links and titles (this may need adjustments based on Google's layout)
        search_results = wd.find_elements(By.CSS_SELECTOR, ".g")
        for result in search_results:
            try:
                link = result.find_element(By.TAG_NAME, "a").get_attribute("href")
                title = result.find_element(By.TAG_NAME, "h3").text
                print(f"Job Title: {title}\nLink: {link}\n")
            except:
                pass  # Skip if any element is not found

    except Exception as e:
        print(f"Error occurred while searching: {e}")

    finally:
        wd.quit()


if __name__ == "__main__":
    # Example usage
    skill_set = ["Python", "Machine Learning", "Data Science"]
    search_jobs_with_skills(skill_set)
```

```python
# prompt: to the above code, add "city or country" preference as input and filter the jobs based out of the given input city or country

import sys
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

!pip install selenium
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin


def search_jobs_with_skills(skills, location):
    """
    Searches for jobs on Google matching the given skillset and location.
    """

    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    wd = webdriver.Chrome(options=chrome_options)

    wd.get("https://www.google.com/")

    search_box = wd.find_element(By.NAME, "q")
    search_query = "jobs " + " ".join(['"' + skill + '"' for skill in skills]) + " in " + location
    search_box.send_keys(search_query)
    search_box.send_keys(Keys.RETURN)

    try:
        # Wait for search results to load
        WebDriverWait(wd, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, ".g"))
        )

        # Extract job links and titles (this may need adjustments based on Google's layout)
        search_results = wd.find_elements(By.CSS_SELECTOR, ".g")
        for result in search_results:
            try:
                link = result.find_element(By.TAG_NAME, "a").get_attribute("href")
                title = result.find_element(By.TAG_NAME, "h3").text
                print(f"Job Title: {title}\nLink: {link}\n")
            except:
                pass  # Skip if any element is not found

    except Exception as e:
        print(f"Error occurred while searching: {e}")

    finally:
        wd.quit()


if __name__ == "__main__":
    # Example usage
    skill_set = ["Python", "Machine Learning", "Data Science"]
    city_or_country = "Chennai" #@param {type:"string"}
    search_jobs_with_skills(skill_set, city_or_country)
```
