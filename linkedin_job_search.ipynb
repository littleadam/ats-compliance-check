```python
# prompt: write a python code using selenium to search jobs in linkedin.com ,using the given skillset as comma separated strings and location as city/country .list the link of top 20 jobs
!pip install selenium
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin


import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


def search_linkedin_jobs(skills, location):
  """
  Searches for jobs on LinkedIn matching the given skillset and location.
  """

  chrome_options = webdriver.ChromeOptions()
  chrome_options.add_argument('--headless')
  chrome_options.add_argument('--no-sandbox')
  chrome_options.add_argument('--disable-dev-shm-usage')
  wd = webdriver.Chrome(options=chrome_options)

  try:
    wd.get("https://www.linkedin.com/jobs/")
    time.sleep(100)  # Allow time for the page to load

    # Find the search bar
    keyword_search_bar = wd.find_element(By.CSS_SELECTOR, "input[aria-label='Search jobs']")
    location_search_bar = wd.find_element(By.CSS_SELECTOR, "input[aria-label='Search by location']")

    # Enter keywords and location
    keyword_search_bar.send_keys(" ".join(skills))
    location_search_bar.send_keys(location)
    location_search_bar.send_keys(Keys.RETURN)

    time.sleep(100)  # Allow time for search results to load

    # Extract job links
    job_links = []
    job_elements = wd.find_elements(By.CSS_SELECTOR, "a[data-tracking-control-name='public_jobs_jserp_job_click']")
    for i, job_element in enumerate(job_elements[:20]):  # Get the first 20 job links
        job_link = job_element.get_attribute("href")
        job_links.append(job_link)

    return job_links

  except Exception as e:
    print(f"Error occurred while searching LinkedIn jobs: {e}")
  finally:
    wd.quit()


if __name__ == "__main__":
    skill_set = ["Python", "Machine Learning", "Data Science"]  #@param {type:"raw"}
    location = "Chennai, Tamil Nadu, India"  #@param {type:"string"}

    job_links = search_linkedin_jobs(skill_set, location)

    if job_links:
        print("Top 20 Job Links:")
        for i, link in enumerate(job_links):
            print(f"{i + 1}. {link}")
```

```python
# prompt: write a python code to extract a profile from linkedin.

!pip install selenium
!apt-get update # to update ubuntu to correctly run apt install
!apt install chromium-chromedriver
!cp /usr/lib/chromium-browser/chromedriver /usr/bin

import sys
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC


def get_linkedin_profile_details(profile_url):
    """
    Extracts details from a LinkedIn profile URL using Selenium.
    """

    chrome_options = webdriver.ChromeOptions()
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--disable-dev-shm-usage')
    wd = webdriver.Chrome(options=chrome_options)

    try:
        wd.get(profile_url)
        time.sleep(5)  # Wait for the page to load completely

        print("Fetching LinkedIn profile details...")
        print(wd)
        # Extract name
        try:
            name = wd.find_element(By.CSS_SELECTOR, 'h1.text-heading-inlarge').text
            print(f"Name: {name}")
        except:
            print("Name not found.")

        # Extract headline
        try:
            headline = wd.find_element(By.CSS_SELECTOR, '.text-body-medium').text
            print(f"Headline: {headline}")
        except:
            print("Headline not found.")

        # Extract location
        try:
            location = wd.find_element(By.CSS_SELECTOR, '.pv-text-details__left-panel').text
            print(f"Location: {location}")
        except:
            print("Location not found.")

        # Extract experience
        try:
            experience_section = wd.find_element(By.ID, 'experience-section')
            experience_list = experience_section.find_elements(By.CSS_SELECTOR, '.pv-entity__summary-info')
            for experience in experience_list:
                title = experience.find_element(By.CSS_SELECTOR, '.pv-entity__summary-info--title').text
                company = experience.find_element(By.CSS_SELECTOR, '.pv-entity__summary-info--company').text
                print(f"Experience: {title} at {company}")
        except:
            print("Experience not found.")

    except Exception as e:
        print(f"Error occurred while fetching LinkedIn profile details: {e}")
    finally:
        wd.quit()


if __name__ == "__main__":
    profile_url = "https://www.linkedin.com/in/aruldswamy"  #@param {type:"string"}
    get_linkedin_profile_details(profile_url)
```
